```python
import numpy as np
class Cube:
    def __init__(self):
        self._u = np.matrix([['ðŸŸ¨', 'ðŸŸ¨'], ['ðŸŸ¨', 'ðŸŸ¨']], dtype = 'str')
        
        self._d = np.matrix([['â¬œ', 'â¬œ'], ['â¬œ', 'â¬œ']], dtype = 'str')
        
        self._f = np.matrix([['ðŸŸ©', 'ðŸŸ©'], ['ðŸŸ©', 'ðŸŸ©']], dtype = 'str')

        self._b = np.matrix([['ðŸŸ¦', 'ðŸŸ¦'], ['ðŸŸ¦', 'ðŸŸ¦']], dtype = 'str')

        self._r = np.matrix([['ðŸŸ§', 'ðŸŸ§'], ['ðŸŸ§', 'ðŸŸ§']], dtype = 'str')

        self._l = np.matrix([['ðŸŸ¥', 'ðŸŸ¥'], ['ðŸŸ¥', 'ðŸŸ¥']], dtype = 'str')
    
    
    def __repr__(self):
        return ('\n\t    ' + str(self._u[0]) + '\n\t    ' + str(self._u[1]) + 
                '\n' + str(self._l[0]) + str(self._f[0]) + str(self._r[0]) + " " + str(self._b[0]) + 
                '\n' + str(self._l[1]) + str(self._f[1]) + str(self._r[1]) + " " + str(self._b[1]) + 
                '\n\t    ' + str(self._d[0]) + '\n\t    ' + str(self._d[1]))

    
    # 2 = double move, 3 = prime move
    def move(self, mv, modifier = 1):
        for i in range(modifier):
            mv()
            
            
    # Rotates only the stickers on a face. Note: impossible on a real cube.  This is used to build other moves.
    def rotate(self, face, modifier = 1):
        for i in range(modifier):
            np.rot90(face) #rotates counter-clockwise
    
    
    # Turns layer 90 degrees clockwise with respect to the face.
    def U(self):
        temp = self._l[0].copy()
        self._l[0] = self._f[0]
        self._f[0] = self._r[0]
        self._r[0] = self._b[0]
        self._b[0] = temp
        self.rotate(self._u)
    
    
    def D(self):
        temp = self._f[1].copy()
        self._f[1] = self._l[1]
        self._l[1] = self._b[1]
        self._b[1] = self._r[1]
        self._r[1] = temp
        self.rotate(self._d)        
        
        
    #broken    
    def F(self):
        self.X()
        self.U()
        self.move(self.X, 3)
        
    
    # Rotates whole cube with respect to R
    def X(self):
        self.rotate(self._r)
        temp = self._u.copy()
        self._u = self._f
        self._f = self._d
        self._d = self._b
        self._b = temp   
        self.rotate(self._l, 3)
    

#     def Y(self):
#         self.rotate(self._u)
#         temp = self._l.copy()
#         self._l = self._f
#         self._f = self._r
#         self._r = self._b
#         self._b = temp   
#         self.rotate(self._d, 3)
    
#     def Z(self):
#         pass
```
