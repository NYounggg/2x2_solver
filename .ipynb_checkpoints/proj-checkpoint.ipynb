```python
import numpy as np
import random
import platform
import itertools as it
import copy

class Cube:
    def __init__(self):
        self._u = np.matrix([['ðŸŸ¨', 'ðŸŸ¨'], ['ðŸŸ¨', 'ðŸŸ¨']], dtype = 'str')      
        self._d = np.matrix([['â¬œ', 'â¬œ'], ['â¬œ', 'â¬œ']], dtype = 'str')     
        self._f = np.matrix([['ðŸŸ©', 'ðŸŸ©'], ['ðŸŸ©', 'ðŸŸ©']], dtype = 'str')
        self._b = np.matrix([['ðŸŸ¦', 'ðŸŸ¦'], ['ðŸŸ¦', 'ðŸŸ¦']], dtype = 'str')
        self._r = np.matrix([['ðŸŸ§', 'ðŸŸ§'], ['ðŸŸ§', 'ðŸŸ§']], dtype = 'str')
        self._l = np.matrix([['ðŸŸ¥', 'ðŸŸ¥'], ['ðŸŸ¥', 'ðŸŸ¥']], dtype = 'str')
        self._movestack = []
    
    
    def __repr__(self):
        # adjust space size in output depending on platform (Linux / Windows)
        if platform.system() == 'Linux':
            spaces = '   '
        else:
            spaces = '    '
            
        outstr = ''
        for i in range(2):
            outstr += spaces
            for j in range(2):
                outstr += self._u[i,j]
            outstr += '\n'
        
        for i in range(2):
            for j in range(2):
                outstr += self._l[i, j]
            for j in range(2):
                outstr += self._f[i, j]
            for j in range(2):
                outstr += self._r[i, j]
                
            outstr += '\t'
            for j in range(2):
                outstr += self._b[i, j]
            outstr += '\n'
            
        for i in range(2):
            outstr += spaces
            for j in range(2):
                outstr += self._d[i,j]
            outstr += '\n'
        
        return outstr

    
    # 2 = double move, 3 = counter-clockwise move
    def move(self, mv, modifier = 1):
        for i in range(modifier):
            mv()
        self._movestack.append((mv, modifier))
        
    
    # Computes a move from the passed mvstring.  mvstring modifier must be an int 1-3
    def move_from_string(self, mvstring):
        modifier = int(mvstring[1])
        mv = None
        if mvstring[0].lower() == 'r':
            mv = self.R
        elif mvstring[0].lower() == 'l':
            mv = self.L
        elif mvstring[0].lower() == 'u':
            mv = self.U
        elif mvstring[0].lower() == 'd':
            mv = self.D
        elif mvstring[0].lower() == 'f':
            mv = self.F
        elif mvstring[0].lower() == 'b':
            mv = self.B
        elif mvstring[0].lower() == 'x':
            mv = self.X
        elif mvstring[0].lower() == 'y':
            mv = self.Y
        elif mvstring[0].lower() == 'z':
            mv = self.Z
            
        for i in range(modifier):
            mv()
            
        self._movestack.append((mv, modifier))
    
    
    # returns stack containing all previous moves performed
    def get_movestack(self):
        return self._movestack
    
    
    # returns string representation of movestack
    def get_movestack_as_str(self):
        outstr = ''
        for i in self._movestack:
            outstr += i[0].__name__ + str(i[1]) + ' '
        return outstr
    
    
    # undoes the previous move and calls pop() on the move stack
    def undo(self):
        self._movestack.pop()
    
    
    # Turns layer 90 degrees clockwise with respect to the face.
    def U(self):
        temp = self._l[0].copy()
        self._l[0] = self._f[0]
        self._f[0] = self._r[0]
        self._r[0] = self._b[0]
        self._b[0] = temp
        self._u = np.rot90(self._u, 3)

        
    def D(self):
        temp = self._f[1].copy()
        self._f[1] = self._l[1]
        self._l[1] = self._b[1]
        self._b[1] = self._r[1]
        self._r[1] = temp
        self._d = np.rot90(self._d, 3)        
        
    
    def F(self):
        temp = self._u[1,:].copy()
        self._u[1,:] = np.flip(np.transpose(self._l[:,1]), axis = 1)
        self._l[:,1] = np.flip(np.transpose(self._d[0,:]), axis = 1)
        self._d[0,:] = np.flip(np.transpose(self._r[:,0]), axis = 1)
        self._r[:,0] = np.flip(np.transpose(temp), axis = 1)
        self._f = np.rot90(self._f, 3)
        
        
    def B(self):
        temp = self._u[0,:].copy()
        self._u[0,:] = np.transpose(self._r[:,1])
        self._r[:,1] = np.flip(np.transpose(self._d[1,:]), axis = 0)
        self._d[1,:] = np.transpose(self._l[:,0])
        self._l[:,0] = np.flip(np.transpose(temp), axis = 0)
        self._b = np.rot90(self._b, 3)
        
        
    def R(self):
        temp = self._u[:,1].copy()
        self._u[:,1] = self._f[:,1]
        self._f[:,1] = self._d[:,1]
        self._d[:,1] = np.flip(self._b[:,0], axis=0)
        self._b[:,0] = temp[::-1]    
        self._r = np.rot90(self._r, 3)
        
        
    def L(self):
        temp = self._d[:,0].copy()
        self._d[:,0] = self._f[:,0]
        self._f[:,0] = self._u[:,0]
        self._u[:,0] = np.flip(self._b[:,1], axis=0)
        self._b[:,1] = temp[::-1]    
        self._l = np.rot90(self._l, 3)
    
    
    #X, Y, Z rotate entire cube
    def X(self):
        temp = self._u.copy()
        self._u = self._f.copy()
        self._f = self._d.copy()
        
        self._d = np.flip(self._b.copy(), axis=0)
        self._d = np.flip(self._d, axis=1)
        self._b = np.flip(temp, axis=0)
        self._b = np.flip(self._b, axis=1)
        
        self._r = np.rot90(self._r, 3)
        self._l = np.rot90(self._l)
        
    
    def Y(self):
        temp = self._f.copy()
        self._f = self._r.copy()
        self._r = self._b.copy()
        self._b = self._l.copy()
        self._l = temp
        
        self._u = np.rot90(self._u, 3)
        self._d = np.rot90(self._d)

        
    def Z(self):
        temp = self._u.copy()
        self._u = np.rot90(self._l.copy(), 3)
        self._l = np.rot90(self._d.copy(), 3)
        self._d = np.rot90(self._r.copy(), 3)
        self._r = np.rot90(temp, 3)
                
        self._f = np.rot90(self._f, 3)
        self._b = np.rot90(self._b)
    
     
    # Generates scramble with no consecutive repeating moves.  
    # R, U, F moves on a 2x2 are enough to fully scramble
    def scramble(self, length):
        print_str = 'Your scramble: (Z2) '
        modifiers =({"1" : "", "2" : "2", "3" : "\'"})
        move = ''  
        prev = ''
        for i in range(length):        
            modifier = random.choice(list(modifiers.keys()))
            while move == prev:
                move = random.choice(['r', 'u', 'f'])
            prev = move
            
            if move == 'r':
                self.move(self.R, int(modifier))
            elif move == 'u':
                self.move(self.U, int(modifier))
            elif move == 'f':
                self.move(self.F, int(modifier))
                
            print_str += move.upper() + modifiers[modifier] + " "
        print(print_str)
        
        # re-initialize move stack
        self._movestack = []
        
        
    # Undoes the previous move.  Rotations cannot be undone as of now
    def undo(self):
        if len(self._movestack) > 0:
            modifier = ''
            to_undo = self._movestack.pop()
            
            if to_undo[1] == 1:
                modifier = '3'
            elif to_undo[1] == 2:
                modifier = '2'
            elif to_undo[1] == 3:
                modifier = '1'

            self.move_from_string(to_undo[0].__name__ + modifier)
            self._movestack.pop() # pop the 'undoing' move
    
    
    # Checks if a specific face is solved
    def face_solved(self, face):
        face_flattened = np.ravel(face)
        if len({i for j in face_flattened for i in j}) == 1:
            return True
        return False
        
    
    # Checks if the R, U, F faces are solved
    def solved(self):
        if (
            self.face_solved(self._u) 
            and self.face_solved(self._r) 
            and self.face_solved(self._f)
        ):
            return True
        return False
    
    
    # Flattens current permutation to 1D string in the format UUUU,DDDD,FFFF,BBBB,RRRR,LLLL
    def to_1d_string(self):
        outstring = ''
        for i in np.ravel(self._u):
            outstring += i
        outstring += ','
        for i in np.ravel(self._d):
            outstring += i
        outstring += ','
        for i in np.ravel(self._f):
            outstring += i
        outstring += ','
        for i in np.ravel(self._b):
            outstring += i
        outstring += ','
        for i in np.ravel(self._r):
            outstring += i
        outstring += ','
        for i in np.ravel(self._l):
            outstring += i
        
        return outstring       
    
    
    # finds solution to current scramble when solution length is known.  Does not solve cube object itself
    def solve(self, solution_len):
        self._movestack = []
        original_state = copy.deepcopy(self)
        
        for i in gen_solution(solution_len):
            for j in i:
                self.move_from_string(j)
            if self.solved():
                return self.get_movestack_as_str()
                break
            else:
                self = copy.deepcopy(original_state)
        
```

```python
def gen_solution(solution_len):
    moves = ['R1', 'U1', 'F1', 'R2', 'U2', 'F2', 'R3', 'U3', 'F3']
    perms = [i for i in it.product(moves, repeat = solution_len)]

    has_consec_dup = False

    for i in range(len(perms)):
        has_consec_dup = False
        for j in range(i):
            if j+1 < len(perms[i]) and perms[i][j][0] == perms[i][j+1][0]:
                has_consec_dup = True
                break

        if not has_consec_dup and not i == 0:
            yield perms[i]
```

```python
from IPython.display import clear_output

class main:   
    
    cube = Cube()
    print(cube)
    
    choice = input("Choose a move >>> ").lower()
    while(not choice == 'quit'):
        if len(choice) == 2:
            if choice[len(choice)-1] == '\'':
                modifier = 3
            elif choice[len(choice)-1] == '2':
                modifier = 2
        elif len(choice) == 1:
            modifier = 1

        if choice[0] == 'u' and len(choice) != 4:
            cube.move(cube.U, modifier)
        elif choice[0] == 'd':
            cube.move(cube.D, modifier)
        elif choice[0] == 'f':
            cube.move(cube.F, modifier)
        elif choice[0] == 'b':
            cube.move(cube.B, modifier)
        elif choice[0] == 'r':
            cube.move(cube.R, modifier)
        elif choice[0] == 'l':
            cube.move(cube.L, modifier)
        elif choice[0] == 'x':
            cube.move(cube.X, modifier)
        elif choice[0] == 'y':
            cube.move(cube.Y, modifier)
        elif choice[0] == 'z':
            cube.move(cube.Z, modifier)
        elif choice == 'scramble':
            cube = Cube()
            cube.scramble(10)      
        elif choice == 'prev':
            printstr = ''
            for i in cube.get_movestack():
                printstr += i[0].__name__ + str(i[1]) + ' '
            print('Previous moves:\n' + printstr)
        elif choice == "undo":
            cube.undo()
                
        # overwrite previous console output
        if not (choice == 'scramble' or choice == 'prev'):
            clear_output()
        
        print(cube)
        if(cube.solved()):
            print('Solved!')
            
        choice = input("Chose a move >>> ").lower()
        
```
